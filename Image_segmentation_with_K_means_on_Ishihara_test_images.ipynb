## Importing libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import files
```

## Testing on number 12

### Kmeans implementation and image processing

```python

# Global variable for k (number of clusters)
k = 5

# K-means implementation from scratch with efficient initialization
def kmeans_from_scratch(X, max_iters=100):
    # Initialize centroids by selecting k random distinct pixels from the image
    unique_pixels = np.unique(X, axis=0)
    centroids = unique_pixels[np.random.choice(unique_pixels.shape[0], k, replace=False)]

    for _ in range(max_iters):
        # Assign each pixel to the nearest centroid
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        # Update centroids by averaging the pixels assigned to each cluster
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])

        # Check for convergence (if centroids do not change)
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids
    return centroids, labels

# Function to process the image and assign clusters to different color channels
def process_image(image_path):
    # Read the image
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB

    # Flatten the image to 2D array for clustering
    pixels = img.reshape(-1, 3)

    # Apply K-means with better initialization from scratch
    centroids, labels = kmeans_from_scratch(pixels)

    return img, labels, img.shape  # Return original image, labels, and shape

uploaded = files.upload()  # Prompt user to upload an image file
image_path = list(uploaded.keys())[0]  # Get the path of the uploaded image

# Step 1: Process the image (find clusters)
original_image, labels, shape = process_image(image_path)
```

### Assigning colours to each cluster and visualization

```python

# Function to assign colors to clusters (in a separate function)
def assign_colors_to_clusters(labels, shape):
    # Reshape labels to match the image dimensions (height, width)
    labels_reshaped = labels.reshape(shape[0], shape[1])

    # Create an empty image for the final output
    clustered_image = np.zeros((shape[0], shape[1], 3), dtype=np.uint8)

    # Assign each cluster to a different channel (R, G, B)
    for i in range(k):

        if i == 2:  # Assign to Blue channel
            clustered_image[labels_reshaped == i] = [255, 0, 0]  # Blue
        else:
          clustered_image[labels_reshaped == i] = [0, 0, 0]  # Black (background)


    return clustered_image

# Show the original image and clustered image side by side
def show_images(original, clustered):
    # Concatenate the original and clustered images side by side for comparison
    concat_image = np.concatenate((original,
                                   193 * np.ones((original.shape[0], int(0.0625 * original.shape[1]), 3), dtype=np.uint8),
                                   cv2.cvtColor(clustered, cv2.COLOR_RGB2BGR)), axis=1)
    plt.imshow(concat_image)
    plt.axis('off')
    plt.show()

# Step 2: Assign colors to clusters
clustered_image = assign_colors_to_clusters(labels, shape)

# Show the results
show_images(original_image, clustered_image)
```
